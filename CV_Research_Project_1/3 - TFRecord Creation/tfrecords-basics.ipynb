Import tfrecord maker

```python
import quick_ml
import os
from quick_ml.tfrecords_maker import create_tfrecord_labeled

from quick_ml.tfrecords_maker import get_addrs_labels

def create_tfrecord(dir, dataset_type):
    addrs, labels = get_addrs_labels(dir)

    image_num = len(os.listdir(f"{dir}\\class1")) + len(os.listdir(f"{dir}\\class2")) + len(os.listdir(f"{dir}\\class3")) + len(os.listdir(f"{dir}\\class4"))

    output_filename = f"D:\\New Datasets\\Full_ExoNet_4C_Final_TFRecord\\{dataset_type}\\224x224-{image_num}.tfrec"

    create_tfrecord_labeled(addrs, labels, output_filename, IMAGE_SIZE = (256,256), num_parts = 1)
```

```python
Dataset_type_list = ['train', 'val', 'test']
# Dataset_type_list = ['test']
# Dataset_folders_list = ['IS', 'IS-LG', 'LG', 'LG-IS']

for type in Dataset_type_list:
    # for folder in Dataset_folders_list:
    DIR = f"D:\\New Datasets\\Project1_Sorted_Dataset_1\\{type}"
    
    create_tfrecord(DIR, type)
```
